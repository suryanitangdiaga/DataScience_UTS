Mempelajari penggunaan Matplotlib dan Seaborn untuk membuat visualisasi data, seperti bar chart, histogram, boxplot, scatter plot, serta menyusun dashboard visualisasi statis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
print("Jumlah Baris dan Kolom:")
print(df.shape)

df.info()
```

```python
avg_bill = df.groupby('day', observed=False)['total_bill'].mean()

plt.figure(figsize=(8,5))
avg_bill.plot(kind='bar')
plt.title('Rata-rata Total Bill per Hari')
plt.ylabel('Total Bill')
plt.xlabel('Hari')
plt.show()
```

### What?
Terdapat hubungan positif antara total bill dan tip.

### So What?
Semakin besar total bill, tip yang diberikan pelanggan cenderung meningkat.

### Now What?
Manajemen dapat menggunakan informasi ini untuk mengevaluasi pola pelayanan dan kepuasan pelanggan.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['total_bill'], kde=True)

plt.title('Distribusi Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frekuensi')

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', data=df)

plt.title('Distribusi Total Bill per Hari')

plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='total_bill', y='tip', data=df)

plt.title('Hubungan Total Bill dan Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.show()
```
