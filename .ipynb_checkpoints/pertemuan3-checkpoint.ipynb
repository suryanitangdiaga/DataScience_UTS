## Membaca Dataset

```python
import pandas as pd

df = pd.read_csv("../housing_dirty.csv")
df.head()
```

## Eksplorasi Awal Dataset

```python
print("INFO DATASET")
df.info()
```

```python
df.describe()
```

```python
print("Jumlah Missing Value Tiap Kolom")
df.isnull().sum()
```

## Menghapus Data Duplikat

```python
print("Jumlah Duplikat Sebelum:")
print(df.duplicated().sum())
```

## Normalisasi String

```python
df['kota'] = df['kota'].str.strip().str.title()

df['kondisi'] = df['kondisi'].str.strip().str.lower()
```

```python
df['kota'] = df['kota'].replace({
    'Jkt':'Jakarta',
    'Mdn':'Medan',
    'Sby':'Surabaya',
    'Dpk':'Depok'
})

df[['kota','kondisi']].head()
```

## Menangani Missing Value

```python
df.isnull().sum()
```

```python
for kolom in df.select_dtypes(include=np.number).columns:
    df[kolom] = df[kolom].fillna(df[kolom].median())
```

```python
for kolom in df.select_dtypes(include='object').columns:
    df[kolom] = df[kolom].fillna(df[kolom].mode()[0])
```

```python
df.isnull().sum()
```

## Menangani Outlier Menggunakan IQR

```python
for kolom in ['harga_juta','luas_m2']:

    Q1 = df[kolom].quantile(0.25)
    Q3 = df[kolom].quantile(0.75)

    IQR = Q3 - Q1

    lower = Q1 - (1.5 * IQR)
    upper = Q3 + (1.5 * IQR)

    df[kolom] = df[kolom].clip(lower, upper)
```

```python
df[['harga_juta','luas_m2']].describe()
```

## Validasi Akhir

```python
print("Total Missing Value")
print(df.isnull().sum().sum())
```

```python
print("Total Duplikat")
print(df.duplicated().sum())
```

## Menyimpan Dataset Bersih

```python
df.to_csv("housing_clean.csv", index=False)

print("Dataset berhasil disimpan")
```
