## Tujuan Pembelajaran

Memahami struktur data Python, penggunaan NumPy untuk komputasi numerik, dan Pandas untuk manipulasi data dalam Data Science.

```python
data = [10, 20, 'Python', True, 3.14]

print(data[0])
print(data[-1])
print(data[1:3])

data.append(99)
data.insert(1, 55)
data.remove(20)

print(data)
```

```python
mahasiswa = {
    'nama': 'Budi Santoso',
    'nim': '2024001',
    'prodi': 'Informatika',
    'ipk': 3.85
}

print(mahasiswa['nama'])
print(mahasiswa.get('usia', 0))

mahasiswa['angkatan'] = 2024
del mahasiswa['nim']

print(mahasiswa)
```

```python
import numpy as np
import pandas as pd
```

```python
a1D = np.array([1, 2, 3, 4, 5])
a2D = np.array([[1,2,3],[4,5,6]])

print(a1D)
print(a2D)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([10,20,30,40,50])

print(a + b)
print(a * b)
print(a ** 2)
```

```python
rng = np.random.default_rng(seed=42)
data = rng.normal(0, 1, 1000)

print("Mean :", np.mean(data))
print("Std  :", np.std(data))
print("Min  :", np.min(data))
print("Max  :", np.max(data))
```

```python
nilai = pd.Series(
    [85,92,78,95],
    index=['Alice','Bob','Caca','Dina']
)

print(nilai['Bob'])
print(nilai.mean())
```

```python
data = {
    'nama':['Ali','Budi','Caca','Dina'],
    'usia':[22,25,21,23],
    'kota':['Jakarta','Bandung','Surabaya','Medan'],
    'ipk':[3.8,3.5,3.9,3.7]
}

df = pd.DataFrame(data)

print(df)
print(df.shape)
print(df.columns)
```

```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv(url)

df.head()
```

```python
df[df['Survived'] == 1].head()
```

```python
df.groupby('Pclass')['Survived'].mean()
```
